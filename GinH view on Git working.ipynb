```python
# Run it in notebook


import plotly.plotly as py
import plotly.graph_objs as go
# a
x_data = df['a']

#b
y_data = df['b']

#c
text = df['c']
    
# Base
#d
trace0 = go.Bar(base=df['d'],
    x=x_data,
#e is actor 2 age                
    y=df['Actor 2 Age'],
    text=df['Actor 2 Name'],
    marker=dict(
        color='rgba(1,1,1, 0.0)',
    )
)
# Revenue
#f is actor 2 age
trace1 = go.Bar(base=df['Actor 2 Age'],
    x=x_data,
#g is age difference
    y=df['Age Difference'],
    text=df['Actor 1 Name'],
    marker=dict(
        color='rgba(55, 128, 191, 0.7)',
        line=dict(
            color='rgba(55, 128, 191, 0.7)',
            width=0.0,
        )
    )
)
data = [trace0, trace1]
layout = go.Layout(
     xaxis=dict(
        tickangle=-45,
        tickfont=dict(
            family='PT Sans Narrow',
            size=12,
            color='black'
        ),
        autorange=True,
        showgrid=False,
        zeroline=False,
        showline=False,
        ticks='',
     ),
    
    title="Hollywood Coupling: Age Discrepancy between Male and Female Leads in Movies Over Time",
    barmode='overlay',
    paper_bgcolor='rgba(245, 246, 249, 1)',
    plot_bgcolor='rgba(245, 246, 249, 1)',
    showlegend=False
)


annotations = []

for i in range(0, 9):
    annotations.append(dict(x=x_data[i], y=y_data[i], text=text[i],
                                  font=dict(family='PT Sans Narrow', size=12,
                                  color='#191919'),
                                  showarrow=False,))
    layout['annotations'] = annotations
fig = go.Figure(data=data, layout=layout)
fig['layout'].update(titlefont=dict(
            family='Old Standard TT',
            size=20,
            color='#7f7f7f'
            ),
            paper_bgcolor='#FFFFFF', plot_bgcolor='#FFFFFF')
pyo.plot(fig, include_plotlyjs=False, output_type='div')

# pyo.iplot(fig)

# div = offplot.plot(fig, show_link=False, output_type="div", include_plotlyjs=False)
# change to
```
